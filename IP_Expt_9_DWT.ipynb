Discrete Wavelet Transform

Aim : To implement Discrete Wavelet Transform on an image and anlayze the LL, LH, HL and HH coefficeints.

Theory: Wavelet transform has recently become a very popular when it comes to analysis, de-noising and compression of signals and images. This section describes functions used to perform single- and multilevel Discrete Wavelet Transforms.<br>
pywt.dwt2 computes the single-level 2-D discrete wavelet transform (DWT) of the input data X using the wname wavelet. dwt2 returns the approximation coefficients matrix cA and detail coefficients matrices cH, cV, and cD (horizontal, vertical, and diagonal, respectively).


```python
# Import Libraries
import cv2
import matplotlib.pyplot as plt
import pywt
```

Image -

h high

2) V edges = LH
 
3) HE = HL
 
4) DE = HH
 
image(100 bits) -> LPF -> LPF = LL (approx)

image -> LPF -> HPF = LH (vertical edge)

image -> HPF -> LPF = HL (horizontal egde)

image -> HPF -> HPF = HH (diagonal edge)

Perform DWT image decomposition and view the LL, LH, HL and HH coefficeints 

```python
# Load image
img = cv2.imread("house.tif",0)

# Wavelet transform of image, and plot approximation and details
LL,(HL,LH,HH) = pywt.dwt2(img,"Haar")

# Plot the coefficients
plt.figure(figsize = (10,10))

plt.subplot(221)
plt.imshow(LL,cmap='gray')
plt.title("LL Coeff")
plt.subplot(222)
plt.imshow(HL,cmap='gray')
plt.title("HL Coeff")
plt.subplot(223)
plt.imshow(LH,cmap='gray')
plt.title("LH Coeff")
plt.subplot(224)
plt.imshow(HH,cmap='gray')
plt.title("HH Coeff")
```

# Compute the DWT of a image and reconstruct the image using IDWT

```python
# Load image
img2 = cv2.imread("house.tif",0)

# Wavelet transform of image, and plot approximation and details
X = pywt.dwt2(img2,"Haar")
recovered_img2 = pywt.idwt2(X,"Haar")

plt.figure(figsize = (10,10))
plt.subplot(121)
plt.imshow(img2,cmap='gray')
plt.title("Original Image")
plt.subplot(122)
plt.imshow(recovered_img2,cmap='gray')
plt.title("Recovered Image")

```
