To convert a color (RGB) image to Grayscale

This example converts an image with RGB channels into an image with a single grayscale channel.

Y = 0.2125 R + 0.7154 G + 0.0721 

These weights are used by CRT phosphors as they better represent human perception of red, green and blue than equal weights

```python
from skimage import io
from skimage.color import rgb2gray
import matplotlib.pyplot as plt

img  = io.imread('cat.png')

#convet color to grayscale
I  = rgb2gray(img)

plt.subplot(1,2,1)  #rows,columns,positions
plt.title("Original Image")
io.imshow(img)

plt.subplot(1,2,2)
plt.title("Modified Image")
io.imshow(I)
```

While converting a color to gray image you can loose some information 
Ishihara plates are used to check color blindness

```python
from skimage import io
from skimage.color import rgb2gray
import matplotlib.pyplot as plt

img2  = io.imread('ishihara.jpg')
#convert color to grayscale
I2  = rgb2gray(img2)

plt.figure(figsize=(8,8))

plt.subplot(1,2,1)
io.imshow(img2)
plt.title('Original Image')

plt.subplot(1,2,2)
io.imshow(I2)
plt.title('GrayScale Image Image')
```

To convert image formats

```python
from skimage.color.colorconv import rgb2gray
img = io.imread('ishihara.jpg')
img_gray = rgb2gray(img)
io.imsave('new_ishihara_gray.tif',img_gray)
```

```python
from skimage.transform import resize
import numpy as np

I1=io.imread("ViratKohli.jpg")/255
I2=io.imread("dhoni.jpg")/255

#Checking size and making sizes of both the image same
I2new = resize(I2,(I1.shape[0],I1.shape[1],3))


i=1
plt.figure(figsize=(15,15))
for alpha in np.linspace(0,1,5):
    IB=(alpha*I1)+((1-alpha)*I2new)
    plt.subplot(1,5,i)
    io.imshow(IB)
    i=i+1

plt.show()
plt.tight_layout()
```
